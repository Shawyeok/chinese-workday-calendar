```python
import datetime
import json
import csv
from chinese_calendar import is_holiday, is_workday
```

```python
def days_of_year(year):
    first_day_in_year = datetime.date(year, 1, 1)
    rows = []
    days_in_year = 365
    if year % 4 == 0:
        days_in_year = 366
    for i in range(0, days_in_year):
        date = first_day_in_year + datetime.timedelta(days=i)
        if is_workday(date):
            rows.append((date, 1))
        else:
            rows.append((date, 0))
    return rows

def generate_csv_of_year(year):
    rows = days_of_year(year)
    with open(f'{year}.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)
```

Generate days of the year for `years.properties`

```python
year = 2026
rows = days_of_year(year)
print(year, '=', ''.join([str(row[1]) for row in rows]))
```

Generate csv data for the year `year.csv`

```python
generate_csv_of_year(2026)
```

Generate workday.json for chart on https://observablehq.com/d/5624093ac436f7ce

```python
rows = []
for year in range(2004, 2027):
    for date, workday in days_of_year(year):
        rows.append({ "Date": str(date), "Workday": workday })

with open(f'workday.json', mode='w') as file:
    file.write(json.dumps(rows))
```
